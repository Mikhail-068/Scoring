```python
from numpy.lib.function_base import average
unique_id_nbki = df['id_nbki'].unique() # Запись в переменную unique_categories уникальных категорий приложений

for i in unique_id_nbki:                                                    # Цикл проходится по всем элементам списка unique_categories, переменная category хранит категорию
  df_2 = df[df['id_nbki'] == i]                                               # df_2 хранит в себе все строки той категории, на которой находится цикл
  num_rows = df_2.shape[0]                                                            # В переменной num_rows хранится количество строк получившегося датафрейма
  
  print('Количество обращений одного клиента "', i, '": ', num_rows, sep='')     # Вывод на экран категории и количества приложений (количества строк)
```
