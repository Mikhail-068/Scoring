```python
#Кол-во пропущенных или NaN значений в каждом столбце
df_bad.isnull().sum()
```

```python
#Кол-во пропущенных или NaN значений в каждом столбце в процентах
missing_percentage = df_bad.isnull().sum() / len(df_bad) * 100
missing_percentage_sorted = missing_percentage.sort_values(ascending=False)
print(missing_percentage_sorted)
```
