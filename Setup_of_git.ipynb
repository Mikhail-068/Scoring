# Создание репозитория и расширение доступа пользователям (для владельца репозитория)

Для добавления соавтора в проект нужно перейти в GitHub, создать репозиторий (во вкладке репозитория - New):

image.png

Задайте название репозитория, описание(опционально), выберете тип (публичный или закрытый), при необходимости - создайте README файл и нажмите внизу кнопку Create repository:

image.png

После чего перейти в репозиторий: Settings
image.png

Выберете Collaborators и нажмите Add people:
image.png

Выберете пользователя которому хотите дать доступ (логин в github/e-mail к которому привязан github аккаунт) и добавьте пользователя:

image.png

# Сохранение Colab блокнота (для пользователей)

Далее - если рабочий репозиторий на git-hub приватный (что чаще всего и будет), то необходимо некоторая настройка:
Перейдите на верхней вкладки в "Инструменты"->"Настройки"->"GitHub"
Нажмите строку - Запросить авторизацию в GitHub
И выставьте галочку напротив "Доступ к личным хранилищам и ресурсам организаций":
image.png

Нажмите кнопку Authorize googlecolab

image.png

Готово! Теперь вы можете добавлять файлы Colab в приватный репозиторий!

Далее выбираете на верхней панели в Google Colab выбираете "Файл"->"Сохранить копию в GitHub", у вас появится следующее окно:


image.png

1.  В этом окне вы выбираете хранилище (это название проекта с которым вы работаете и к которым у вас есть доступ, если в списке нет проекта - проверьте дал ли вам доступ владелец репозитория)
2.  Указываете путь к файлу (обычно указываются фамилии в качестве папок - т.е. в строке путь к файлу в данном примере был бы <ваше_имя_фамилия>/setup_of_git.ipynb - т.е. название вашего блокнота в colab)
3. Нажимаете Ок



Теперь в вашем репозитории должен появиться файл colab:
image.png



Этот файл можно будет открыть в colab также:
image.png

Обратите внимание - открыв блокнот из GitHub вы не сможете сохранить изменения, вам нужно будет сохранить его на Google Диск и уже потом изменять, но точно так же при сохранении заново на GitHub - старый Colab блокнот там перезапишется, при изменении имени Colab блокнота - создастся новый в GitHub репозитории.

Следите за порядком в репозитории и сохраняйте свои наработки в своих папках, чтобы никто не запутался :)
