```
# "date of birth", "gender", "education", "position",

print("Группировка данных по пользователям среднее значенние")
df_groupbyManMean = data.groupby(['id_nbki']).agg('mean')
df_groupbyManMean = df_groupbyManMean[[
                              "amount", "psk", "balance", "term_indebtedness_body_loan", "term_indebtedness_interest", "term_indebtedness_other",
            "overdue_indebtedness_interest", "overdue_indebtedness_other", "maximum_amount_overdue_indebtedness", "amount_principal_debt_repaid"]]
df_groupbyManMean
```

```
print("Группировка данных по пользователям сумма кредитов")
df_groupbyManSum = data.groupby(['id_nbki']).agg('sum')
df_groupbyManSum = df_groupbyManSum[[
                              "amount", "psk", "balance", "term_indebtedness_body_loan", "term_indebtedness_interest", "term_indebtedness_other",
            "overdue_indebtedness_interest", "overdue_indebtedness_other", "maximum_amount_overdue_indebtedness", "amount_principal_debt_repaid"]]
df_groupbyManSum
```

```
print("Группировка данных по Образованию и должности средние значения")
df_groupbyEdnPosMean = data.groupby(['education', 'position']).agg('mean')
df_groupbyEdnPosMean = df_groupbyEdnPosMean[[
                              "amount", "psk", "balance", "term_indebtedness_body_loan", "term_indebtedness_interest", "term_indebtedness_other",
            "overdue_indebtedness_interest", "overdue_indebtedness_other", "maximum_amount_overdue_indebtedness", "amount_principal_debt_repaid"]]
df_groupbyEdnPosMean
```

```
#для вывода в формате бе е-7
pd.options.display.float_format ='{:,.3f}'.format

print("Группировка данных по Образованию и должности общие")
df_groupbyEdnPosMean = data.groupby(['education', 'position']).agg('sum')
df_groupbyEdnPosMean = df_groupbyEdnPosMean[[
                              "amount", "psk", "balance", "term_indebtedness_body_loan", "term_indebtedness_interest", "term_indebtedness_other",
            "overdue_indebtedness_interest", "overdue_indebtedness_other", "maximum_amount_overdue_indebtedness", "amount_principal_debt_repaid"]]
df_groupbyEdnPosMean
```
