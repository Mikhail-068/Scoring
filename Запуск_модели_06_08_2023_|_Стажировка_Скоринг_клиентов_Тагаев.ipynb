```python
from IPython.display import display
from keras.utils import to_categorical
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
import pandas as pd # Библиотека для работы с датафреймом
from tensorflow.keras import utils
import numpy as np # Библиотека для работы с массивами
import matplotlib.pyplot as plt # Будем визулизировать данные этой библиотекой
from datetime import datetime
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
pd.set_option("display.max_columns", 100)
```

```python
df_train = pd.read_csv('/content/gdrive/MyDrive/Стажировка Скоринг/Red/db_train_2021_all.csv')
df_train.shape
df_train
```

```python
df_val = pd.read_csv('/content/gdrive/MyDrive/Стажировка Скоринг/Red/db_val_2021_all.csv')
df_val.shape
```

```python
df_test = pd.read_csv('/content/gdrive/MyDrive/Стажировка Скоринг/Red/db_test_2021_all.csv')
df_test.shape
```

```python
df_train = df_train.sample(frac=1)
df_train
```

```python
print(np.unique(df_train['Client_class']))
```

```python
df_val = df_val.sample(frac=1)
```

```python
df_test = df_test.sample(frac=1)
```

```python
# выделение признаков и целевой переменной
X = df_train.drop('Client_class', axis=1)
y = df_train['Client_class']
X_val = df_val.drop('Client_class', axis=1)
y_val = df_val['Client_class']
X_test = df_test.drop('Client_class', axis=1)
y_test = df_test['Client_class']
```

```python
print(np.unique(y))
```

```python
# Преобразование ответов в формат one_hot_encoding
y = to_categorical(y, num_classes = 4)
y_val = to_categorical(y_val, num_classes = 4)
y_test = to_categorical(y_test, num_classes = 4)
```

```python
from sklearn.preprocessing import StandardScaler # Масштабирование данных
```

```python
# Нормализуем данные
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
scaler.fit(X_val)
X_val = scaler.transform(X_val)
scaler.fit(X_test)
X_test_ = scaler.transform(X_test)
print(X) # Посмотрим на нормализованные данные
print(np.unique(y))
```

```python
import autokeras as ak
```

```python
search = ak.AutoModel(
    inputs=[ak.StructuredDataInput()],
    outputs=[ak.ClassificationHead(loss="categorical_crossentropy", metrics=["accuracy"])],
    objective="accuracy",
    tuner="hyperband",
    overwrite=True,
    max_trials=10,
)
search.fit(X,                              # Обучающая выборка
           y,                              # Метки классов обучающей выборки
           epochs=20,                      # Количество эпох
           batch_size=512,                 # Размер подвыборки для одного шага по данным на эпохе
           validation_data=(X_val, y_val), # Проверочная выборка и метки классов проверочной выборки)
           verbose=1)
```

```python
search.save('/content/gdrive/MyDrive/Стажировка Скоринг/model_search.keras')
```

```python
loss, acc = search.evaluate(X_test, y_test, verbose=0)
```
